***Embedding Model***
- convert a sentence into a vector
- very important for document search and RAG

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings


embeddings=GoogleGenerativeAIEmbeddings(model='models/gemini-embedding-001')
vec=embeddings.embed_query('what is the capital of india')

len(vec)
```

```python
doc=['the capital of india is Delhi', 'The Economical capital is mumbai']
vecs=embeddings.embed_documents(doc)
len(vecs)
```
