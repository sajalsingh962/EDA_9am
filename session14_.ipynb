# [<if out> <if cond> else <else out> <for loop>]


```python
Input=[20,21,30,31,34]
out=[]

#[<output> <for loop>]
[<out> <for loop> <if cond>]


# Rule-1: there is no elif
            # else if
# Rule-2: if after else
# Rule-3: else after output
```

```python
Input=[50,70,91]
out=[]
for i in Input:
    if i>90:
        out.append('A')
    elif i>=60:
        out.append('B')
    else:
        out.append('C')
out
```

```python
Input=[50,70,91]
out=[ 'A' if i>90 \
      else 'B' if i>60 \
     else 'C' for i in Input]
out
```

```python
Input=[1, 0, -1,-2,2, 0, -3,3, 0, -4,4]

neg=[i  for i in Input if i<0]
pos=[i for i in Input if i>0]
zero=[i for i in Input if i==0]
neg,zero, pos
```
