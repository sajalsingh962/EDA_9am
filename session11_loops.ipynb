```python
str1='ola ola ola'
# count the number of a
# idea
# step-1: iterate the str1 using for loop
# step-2: call each letter
# step-3: if that is letter equal toa
# step-4: then update the counter count=count+1
```

```python
str1='ola ola ola'
n=len(str1)
count=0
for i in range(n):
    print(i, str1[i])
    if str1[i]=='a':
        count=count+1
        
print('==answer==')
print(count)
```
