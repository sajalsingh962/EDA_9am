1. what are embeddings?
   - Embeddings are numerical reprensentations of text, audio, or any data. For text (usually our focus), an embedding is:
   - - A vector( a long list of numbers)
     - Generated by a model
     - Designed so that similar meanings -> similar numbers
2. Why Do we Use Embeddings?
    - Because computers don't understand text directrly -- they understand numbers
  
    - Embeddings allow systems to:
  
    - - Search semantically ( meaning based search)
      - Build recommentation systems)
      - measures similarly between text
      - Cluster documentents
      - RAG(retrival-Argumented Generation) Including ChatGroq workflow


3. What is a Similarity Score?
   - Once you have embeddings, you meansure similarity using a mathathical formula
   - The most common one is:

 ##Cosine Similarity
   - 1.0 -> identical meaning
   - 0.0 -> unrelated
   - 1.0 -> opposite meaning
       

cos_sim=(A.B)/(||A|| .||B||)

```python
from sentence_transformers import SentenceTransformer
import numpy as np
```

```python
# Open SOurce model No api - key
# For testing purpose

model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
sentences = [
    'I like apples.',
    'I enjoy eating apples.',
    'I drive a motorcycle.'
]
```

```python
embeddings = model.encode(sentences)
```

```python
def cosine_sim(a, b):
    return np.dot(a, b)/(np.linalg.norm(a) * np.linalg.norm(b))
```

```python
v1=embeddings[0] #I like apples.
v2=embeddings[1] # I enjoy eating apples.
v3=embeddings[2] # I drive a motorcycle.

cosine_sim(v1, v2)
```

```python
Nr=np.dot(v1, v2)
norm_v1=np.linalg.norm(v1)
norm_v2=np.linalg.norm(v2)
Dr=norm_v1*norm_v2
cosine_sim(v1, v2), Nr/Dr
```
