```python

def Sum(n):
    if n==0:
        return 0
    ans=n+Sum(n-1)
    
    return ans
    

Sum(5)


# explaination:
#   ans=5+Sum(5-1) => 5+Sum(4) 
#       => 5+Sum(4-1) => 5+4+Sum(3) 
#                  => 5+Sum(3-1) =>5+4+3+Sum(2) 
#                         => 5+Sum(2-1) =>5+4+3+2+Sum(1) 
#                                => 5+Sum(1-1) =>5+4+3+2+1+Sum(0)
#                                                => 5+4+3+2+1+0 =15
#

```
