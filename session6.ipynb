```python
# wap ask the user take the radius find the area of circle
    # pi=3.14
    # r=100
   # area_of_circle=pi*r*r
r=eval(input('enter radius of circle:'))
pi=3.14
area_of_circle=pi*r*r
print(f'area of circle: {round(area_of_circle,2)}')
```

```python
# wap ask the user take the salary, percetage of tax calculate tax amount
sal=eval(input('enter salary:'))
tax_per=eval(input('enter tax percentage:'))
net_sal=round(sal-(sal*tax_per/100),2)
print(f'salary after tax deduction: {net_sal}')
```
