```python
lables=visa_df['continent'].unique()

#certufed count list=[]
cert_count_list=[]
# denited count list=[]
denied_count_list=[]
for i in lables:
    con1=visa_df['continent']==i
    con2=visa_df['case_status']=='Certified'
    con3=visa_df['case_status']=='Denied'
    cert_con=con1&con2
    den_cert=con1& con3
    # append(len(visa_df[cert_con]))
    cert_count_list.append(len(visa_df[cert_con]))
    denied_count_list.append(len(visa_df[den_cert]))

cert_count_list, denied_count_list
#create DataFrame
cols=sorted(visa_df['case_status'].unique())
print(cols)
pd.DataFrame(zip(cert_count_list, denied_count_list), columns=cols)
```
